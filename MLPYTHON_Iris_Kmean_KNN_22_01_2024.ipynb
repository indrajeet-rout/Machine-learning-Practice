```python
import pandas as pd
df=pd.read_csv("/content/datafiles/iris.csv")
```

```python
df_feature=df.iloc[:,0:4]
```

```python
df_feature_final=df_feature.values
```

```python
type(df_feature_final)
```

```python
df_target=df["Targetnames"]
```

```python
df_target
```

```python
df_feature_final=df_feature.values
```

```python
from sklearn.model_selection import train_test_split
```

```python
#train_x,test_x,train_target,test_target=train_test_split(df_feature_final,df_target,test_size=0.2,random_state=45)
```

```python
train_x,test_x,train_target,test_target=train_test_split(df_feature_final,df_target,test_size=0.2,random_state=45)
```

```python
train_x
```

```python
train_target
```

```python
test_x
```

```python
# training the model on training set
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=3)
model.fit(train_x, train_target)
```

```python
# making predictions on the testing set
y_pred = model.predict(test_x)
print(y_pred)
```

```python
print(test_target)
```

```python
# comparing actual response values (y_test) with predicted response values (y_pred)
from sklearn import metrics

print("kNN model accuracy:", metrics.accuracy_score(test_target, y_pred))
```

```python
metrics.confusion_matrix(test_target,y_pred)
```

```python
import pickle
with open("model_june2023.pkl",'wb') as model_pkl:
    pickle.dump(model,model_pkl)
```

```python
import pickle
with open('model_june2023.pkl','rb') as model_file:
    model_suhas=pickle.load(model_file)
```
