```python
df = pd.read_csv("/content/homeprices-pune.csv")
df
```

```python
dummies = pd.get_dummies(df.location)
dummies
```

```python
merged_pune = pd.concat([df,dummies],axis='columns')
merged_pune
```

```python
final = merged_pune.drop(['location'], axis='columns')
final
```

```python
X = final.drop('price', axis='columns')
X
```

```python
y=final.price
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
yhat=model.predict(X)
```

```python
model.score(X,y)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
import numpy as np
# predicting the accuracy score
score=r2_score(y,yhat)
print('r2 score is',score)
print('mean_sqrd_error is==',mean_squared_error(y,yhat))
print('root_mean_squared error of is==',np.sqrt(mean_squared_error(y,yhat)))
```
